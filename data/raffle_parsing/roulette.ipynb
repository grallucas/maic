```python
import pandas as pd
```

```python
# Load the file data\2023 MAIC End-of-Semester Raffle(1-43).xlsx into a Pandas DataFrame (code it)
df = pd.read_excel('raffle_dfs/2023 MAIC End-of-Semester Raffle(1-43).xlsx')

# Grab only the last two columns of the DataFrame
df = df.iloc[:, -2:]
df
```

### Create a list that is Easy to Input Into Online Roulette Wheel
https://wheelofnames.com/

```python
# Given the dataframe, create a new dataframe where a name is repeated x number of rows, where x is the number denoted in the second column. If the number is NaN or 0, then inlcude the person only once

# Create a new dataframe
entries = []

# Iterate through each row in the dataframe

for index, row in df.iterrows():
    # Get the name and number of times to repeat
    name = row[-2]
    num = row[-1]
    
    # If the number is NaN or 0, then include the person only once
    if pd.isnull(num) or num == 0:
        entries.append(name)
    else:
        # Repeat the person x number of times
        for i in range(int(num) + 1):
            entries.append(name)

entries
```

### Better DF Format
Get the df into a format which is easy to remove the points from the online leaderboard

```python
# For df, replace all Nan values with 0 (Free point)
df = df.fillna(0)

# Rename the first column to 'Name' and the second column to 'Points'
df.columns = ['Name', 'Points']
df

# For each name, get rid of white space surrounding the name and title the string
df['Name'] = df['Name'].str.strip().str.title()
df
```

```python
user_data = pd.read_csv('../../data/User_Data.csv')
user_data
```

```python
# Given the amount of points associated with each Name in df, remove those points from user_data 'Current Points'
for index, row in df.iterrows():
    name = row['Name']
    points = row['Points']
    
    # Find the index of the name in user_data
    user_index = user_data[user_data['User'] == name]

    # If user index empty, skip
    if user_index.empty:
        continue

    # Get the index of the user
    user_index = user_index.index[0]
    
    # Subtract the points from the 'Current Points' column
    user_data.loc[user_index, 'Current Points'] = user_data.loc[user_index, 'Current Points'] - points

user_data[user_data['User'] == 'Christian Zastrow']
```

```python
user_data.to_csv('../../data/User_Data.csv', index=False)
```
